```python
import numpy as np
import cv2

img = cv2.imread('miau.jpeg')

# Indicando el factor de escala (mitad de tamaño)
res = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_CUBIC)

# Indicando manualmente el nuevo tamaño deseado de la imagen
#height, width = img.shape[:2] #obtiene los dos primeros valores del w y el h
#res = cv2.resize(img, (2*width, 2*height), interpolation=cv2.INTER_CUBIC) #redimenciona al doble

# Mostrar la imagen
cv2.imshow('Redimensionalizacion', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
 
img = cv2.imread('miau.jpeg',0)
rows,cols = img.shape
 
M = np.float32([[1,0,210],[0,1,20]])
dst = cv2.warpAffine(img,M,(cols,rows))
 
cv2.imshow('Traslacion',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('miau.jpeg',0)
rows,cols = img.shape
 
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('Rotacion',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt #carga la librería para graficar
import cv2
 
img = cv2.imread('miau.jpeg')
rows,cols,ch = img.shape
 
pts1 = np.float32([[100,400],[400,100],[100,100]])
pts2 = np.float32([[50,300],[400,200],[80,150]])
 
 
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.suptitle('Transformación Afín') 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
 
img = cv2.imread('miau.jpeg')
rows,cols,ch = img.shape
 
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
 
M = cv2.getPerspectiveTransform(pts1,pts2)
 
dst = cv2.warpPerspective(img,M,(300,300))
plt.suptitle('Transformación de Perspectiva')  
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
